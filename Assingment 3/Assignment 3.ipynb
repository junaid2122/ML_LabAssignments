```python

# Set style
sns.set(style="whitegrid")
```

```python
# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle("Data Visualizations", fontsize=16)


# Scatter Plot: Age vs. Spending Score
sns.scatterplot(x=df["Age"], y=df["Spending Score (1-100)"], hue=df["Gender"], alpha=0.6, ax=axes[0, 0])
axes[0, 0].set_title("Scatter Plot: Age vs. Spending Score")

# Bar Plot: Average Annual Income by Profession
avg_income = df.groupby("Profession")["Annual Income ($)"].mean().sort_values()
avg_income.dropna().plot(kind="barh", ax=axes[0, 1], color="skyblue")
axes[0, 1].set_title("Bar Plot: Average Annual Income by Profession")

# Box Plot: Distribution of Annual Income
sns.boxplot(y=df["Annual Income ($)"], ax=axes[1, 0], color="lightcoral")
axes[1, 0].set_title("Box Plot: Distribution of Annual Income")

# Pie Chart: Distribution of Genders
gender_counts = df["Gender"].value_counts()
axes[1, 1].pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%", colors=["lightblue", "pink"])
axes[1, 1].set_title("Pie Chart: Gender Distribution")

# Line Chart: Age vs. Spending Score Trend
df_sorted = df.sort_values("Age")
sns.lineplot(x=df_sorted["Age"], y=df_sorted["Spending Score (1-100)"], ax=axes[2, 0], color="purple")
axes[2, 0].set_title("Line Chart: Age vs. Spending Score")

# Hide the last empty subplot
axes[2, 1].axis("off")

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
